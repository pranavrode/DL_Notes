## What is deep learning? Machine learning vs. Deep learning

- **Deep learning** is a subfield of machine learning that focuses on the use of neural networks with multiple layers (deep neural networks) to model and solve complex problems. 
- It is a subset of artificial intelligence (AI) and aims to mimic the human brain's ability to learn and make decisions. 
- In deep learning, these neural networks, often referred to as artificial neural networks (ANNs), consist of interconnected nodes arranged in layers. 
- Each node performs a specific function and processes information received from the previous layer, allowing the network to learn hierarchical representations of data.

- **Machine learning**, on the other hand, is a broader concept that encompasses various algorithms and techniques that enable systems to learn patterns and make predictions or decisions without explicit programming. 
- It involves the development of models that can generalize from data and improve their performance over time based on experience.

- The main distinction between **machine learning and deep learning** lies in the architecture and depth of the models used. - - While traditional machine learning models may involve a **single layer or a limited number of layers**, deep learning models, particularly deep neural networks, consist of **multiple hidden layers**, allowing them to automatically learn intricate features and representations from the data.

- In summary, deep learning is a specialized subset of machine learning that emphasizes the use of deep neural networks with multiple layers to achieve advanced learning and decision-making capabilities, particularly in tasks involving large and complex datasets.
